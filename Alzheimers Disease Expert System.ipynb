```python
#CODE 1 TESTING!!

# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
import numpy as np

# Load the dataset (replace 'alzheimers_disease_dataset.csv' with the correct path)
data = pd.read_csv('D:\\Downloads\\alzheimers_disease_dataset.csv')

# Replace hypothetical 'XXXConfid' with numerical value for classification
data['Diagnosis'] = data['Diagnosis'].replace({'XXXConfid': 0})  # Assuming 0 is no disease, adjust accordingly

# Select only relevant columns for classification (dropping PatientID and DoctorInCharge)
features = data.drop(columns=['PatientID', 'Diagnosis', 'DoctorInCharge'])
target = data['Diagnosis']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Standardize the dataset (scaling)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create a K-Nearest Neighbors (KNN) classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

# Function to ask questions and get user input
def get_user_input():
    print("Please answer the following questions about your health conditions:")
    
    # Asking user input for each feature in the dataset
    user_data = []
    
    age = int(input("Enter your age: "))
    user_data.append(age)
    
    gender = int(input("Enter your gender (0 for Male, 1 for Female): "))
    user_data.append(gender)
    
    ethnicity = int(input("Enter your ethnicity (0-3): "))
    user_data.append(ethnicity)
    
    education_level = int(input("Enter your education level (0-3): "))
    user_data.append(education_level)
    
    bmi = float(input("Enter your BMI (Body Mass Index): "))
    user_data.append(bmi)
    
    smoking = int(input("Do you smoke? (0 for No, 1 for Yes): "))
    user_data.append(smoking)
    
    alcohol_consumption = float(input("How much alcohol do you consume per week (in units): "))
    user_data.append(alcohol_consumption)
    
    physical_activity = float(input("How often do you engage in physical activity (scale 1-10): "))
    user_data.append(physical_activity)
    
    diet_quality = float(input("Rate your diet quality (scale 1-10): "))
    user_data.append(diet_quality)
    
    sleep_quality = float(input("Rate your sleep quality (scale 1-10): "))
    user_data.append(sleep_quality)
    
    family_history = int(input("Do you have a family history of Alzheimer's (0 for No, 1 for Yes): "))
    user_data.append(family_history)
    
    cardiovascular_disease = int(input("Do you have cardiovascular disease? (0 for No, 1 for Yes): "))
    user_data.append(cardiovascular_disease)
    
    diabetes = int(input("Do you have diabetes? (0 for No, 1 for Yes): "))
    user_data.append(diabetes)
    
    depression = int(input("Do you have depression? (0 for No, 1 for Yes): "))
    user_data.append(depression)
    
    head_injury = int(input("Have you ever experienced a head injury? (0 for No, 1 for Yes): "))
    user_data.append(head_injury)
    
    hypertension = int(input("Do you have hypertension? (0 for No, 1 for Yes): "))
    user_data.append(hypertension)
    
    systolic_bp = int(input("Enter your systolic blood pressure (mmHg): "))
    user_data.append(systolic_bp)
    
    diastolic_bp = int(input("Enter your diastolic blood pressure (mmHg): "))
    user_data.append(diastolic_bp)
    
    cholesterol_total = float(input("Enter your total cholesterol level (mg/dL): "))
    user_data.append(cholesterol_total)
    
    cholesterol_ldl = float(input("Enter your LDL cholesterol level (mg/dL): "))
    user_data.append(cholesterol_ldl)
    
    cholesterol_hdl = float(input("Enter your HDL cholesterol level (mg/dL): "))
    user_data.append(cholesterol_hdl)
    
    cholesterol_triglycerides = float(input("Enter your triglycerides level (mg/dL): "))
    user_data.append(cholesterol_triglycerides)
    
    mmse = float(input("Enter your Mini-Mental State Examination (MMSE) score (scale 0-30): "))
    user_data.append(mmse)
    
    functional_assessment = float(input("Rate your functional abilities (scale 1-10): "))
    user_data.append(functional_assessment)
    
    memory_complaints = int(input("Do you experience memory complaints? (0 for No, 1 for Yes): "))
    user_data.append(memory_complaints)
    
    behavioral_problems = int(input("Do you have any behavioral problems? (0 for No, 1 for Yes): "))
    user_data.append(behavioral_problems)
    
    adl = float(input("Rate your ability to perform activities of daily living (ADL) (scale 1-10): "))
    user_data.append(adl)
    
    confusion = int(input("Do you experience confusion? (0 for No, 1 for Yes): "))
    user_data.append(confusion)
    
    disorientation = int(input("Do you experience disorientation? (0 for No, 1 for Yes): "))
    user_data.append(disorientation)
    
    personality_changes = int(input("Do you experience personality changes? (0 for No, 1 for Yes): "))
    user_data.append(personality_changes)
    
    difficulty_completing_tasks = int(input("Do you have difficulty completing tasks? (0 for No, 1 for Yes): "))
    user_data.append(difficulty_completing_tasks)
    
    forgetfulness = int(input("Do you experience forgetfulness? (0 for No, 1 for Yes): "))
    user_data.append(forgetfulness)
    
    # Convert user data to a DataFrame and scale it
    user_df = pd.DataFrame([user_data], columns=features.columns)
    user_scaled = scaler.transform(user_df)
    
    return user_scaled

# Function to predict the probability of Alzheimer's disease
def predict_disease():
    user_input = get_user_input()
    
    # Predict the user's risk of Alzheimer's disease
    prediction = knn.predict(user_input)
    probability = knn.predict_proba(user_input)[0][1] * 100
    
    if prediction == 1:
        print(f"Based on your input, you have a {probability:.2f}% chance of having Alzheimer's disease.")
    else:
        print(f"Based on your input, you have a {probability:.2f}% chance of not having Alzheimer's disease.")

# Run the prediction system
predict_disease()
```

```python
#CODE 2 TESTING! WHICH IS OUR FINAL CODE!!

# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

# Load the dataset (replace 'alzheimers_disease_dataset.csv' with the correct path)
data = pd.read_csv('D:\\Downloads\\alzheimers_disease_dataset.csv')

# Replace hypothetical 'XXXConfid' with numerical value for classification
data['Diagnosis'] = data['Diagnosis'].replace({'XXXConfid': 0})  # Adjust accordingly

# Select only relevant columns for classification (dropping PatientID and DoctorInCharge)
features = data.drop(columns=['PatientID', 'Diagnosis', 'DoctorInCharge'])
target = data['Diagnosis']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Standardize the dataset (scaling)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Hyperparameter tuning using GridSearchCV
param_grid = {'n_neighbors': np.arange(1, 21)}
knn = KNeighborsClassifier()
knn_cv = GridSearchCV(knn, param_grid, cv=5)  # 5-fold cross-validation
knn_cv.fit(X_train_scaled, y_train)

# Best KNN model
best_knn = knn_cv.best_estimator_

# Evaluate the model
y_pred = best_knn.predict(X_test_scaled)
print(classification_report(y_test, y_pred))
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")

# Function to ask questions and get user input
def get_user_input():
    print("Please answer the following questions about your health conditions:")
    
    user_data = []
    
    # Collect user input for each feature
    user_data.append(int(input("Enter your age: ")))
    user_data.append(int(input("Enter your gender (0 for Male, 1 for Female): ")))
    user_data.append(int(input("Enter your ethnicity (0-3): ")))
    user_data.append(int(input("Enter your education level (0-3): ")))
    user_data.append(float(input("Enter your BMI (Body Mass Index): ")))
    user_data.append(int(input("Do you smoke? (0 for No, 1 for Yes): ")))
    user_data.append(float(input("How much alcohol do you consume per week (in units): ")))
    user_data.append(float(input("How often do you engage in physical activity (scale 1-10): ")))
    user_data.append(float(input("Rate your diet quality (scale 1-10): ")))
    user_data.append(float(input("Rate your sleep quality (scale 1-10): ")))
    user_data.append(int(input("Do you have a family history of Alzheimer's (0 for No, 1 for Yes): ")))
    user_data.append(int(input("Do you have cardiovascular disease? (0 for No, 1 for Yes): ")))
    user_data.append(int(input("Do you have diabetes? (0 for No, 1 for Yes): ")))
    user_data.append(int(input("Do you have depression? (0 for No, 1 for Yes): ")))
    user_data.append(int(input("Have you ever experienced a head injury? (0 for No, 1 for Yes): ")))
    user_data.append(int(input("Do you have hypertension? (0 for No, 1 for Yes): ")))
    user_data.append(int(input("Enter your systolic blood pressure (mmHg): ")))
    user_data.append(int(input("Enter your diastolic blood pressure (mmHg): ")))
    user_data.append(float(input("Enter your total cholesterol level (mg/dL): ")))
    user_data.append(float(input("Enter your LDL cholesterol level (mg/dL): ")))
    user_data.append(float(input("Enter your HDL cholesterol level (mg/dL): ")))
    user_data.append(float(input("Enter your triglycerides level (mg/dL): ")))
    user_data.append(float(input("Enter your MMSE score (scale 0-30): ")))
    user_data.append(float(input("Rate your functional abilities (scale 1-10): ")))
    user_data.append(int(input("Do you experience memory complaints? (0 for No, 1 for Yes): ")))
    user_data.append(int(input("Do you have any behavioral problems? (0 for No, 1 for Yes): ")))
    user_data.append(float(input("Rate your ability to perform ADL (scale 1-10): ")))
    user_data.append(int(input("Do you experience confusion? (0 for No, 1 for Yes): ")))
    user_data.append(int(input("Do you experience disorientation? (0 for No, 1 for Yes): ")))
    user_data.append(int(input("Do you experience personality changes? (0 for No, 1 for Yes): ")))
    user_data.append(int(input("Do you have difficulty completing tasks? (0 for No, 1 for Yes): ")))
    user_data.append(int(input("Do you experience forgetfulness? (0 for No, 1 for Yes): ")))
    
    # Convert user data to a DataFrame and scale it
    user_df = pd.DataFrame([user_data], columns=features.columns)
    user_scaled = scaler.transform(user_df)
    
    return user_scaled

# Function to predict the probability of Alzheimer's disease
def predict_disease():
    user_input = get_user_input()
    
    # Predict the user's risk of Alzheimer's disease
    prediction = best_knn.predict(user_input)
    probability = best_knn.predict_proba(user_input)[0][1] * 100
    
    if prediction == 1:
        print(f"Based on your input, you have a {probability:.2f}% chance of having Alzheimer's disease.")
    else:
        print(f"Based on your input, you have a {probability:.2f}% chance of not having Alzheimer's disease.")

# Run the prediction system
predict_disease()
```
